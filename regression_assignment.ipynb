5. What is the method of least squares in linear regression?

* Method of Least Squares (in Linear Regression)

* The method of least squares is a mathematical technique used in linear regression to find the best-fit line for the data.

* It works by: Minimizing the sum of squared errors (residuals)

* The goal is to choose values of

* Œ≤‚ÇÄ (intercept)

* Œ≤‚ÇÅ (slope) such that this total squared error is as small as possible.

7. Name and briefly describe three common evaluation metrics for regression
models.

* Mean Absolute Error (MAE): MAE measures the average absolute difference between the actual values and the predicted values.
It tells how much error the model makes on average, in the same units as the target variable.

* Mean Squared Error (MSE): MSE measures the average of squared errors between actual and predicted values.
Squaring gives more weight to larger errors, so MSE penalizes big mistakes more heavily.

* Root Mean Squared Error (RMSE): RMSE is the square root of MSE.
It represents error in the same units as the target, making it easier to interpret. RMSE is very sensitive to large errors.

8. What is the purpose of the R-squared metric in regression analysis?

* R-squared (R¬≤) is a metric that measures how well a regression model explains the variation in the dependent variable (Y).

* It tells you what percentage of the total variation in Y is explained by the model.

* ùëÖ2 = 0 ‚Üí Model explains none of the variation

* R2 = 1 ‚Üí Model explains all the variation perfectly


```python
# Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.
# answer
import numpy as np
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([3, 4, 2, 5, 6])

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Print slope and intercept
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```
